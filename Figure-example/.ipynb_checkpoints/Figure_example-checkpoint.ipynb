## Basic plot example

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.random.uniform(0, 10, 30)
x.sort()
yt = 2 * x + 1
y = yt + np.random.normal(loc=0, scale=2, size=yt.size)

plt.plot(x, yt, label="model")
plt.plot(x, y, "o")
plt.xlabel("x values (unit)")
plt.ylabel("y values (unit)")
plt.title("A plot")
````

Con la siguiente linea de código pueden cargar el estilo de la figura. 

## Plot example using mplstyle

````python
plt.style.use('https://raw.githubusercontent.com/RAHXION/Resources/main/rahxion_figure_style.mplstyle')
````

````python
fig, ax = plt.subplots()
ax.plot(x, yt, label="Modelo",color='r')
ax.scatter(x, y,marker='o',color='k',label = 'Puntos medidos')
ax.set_xlabel('x values ($m^3$)')
ax.set_ylabel("y values ($\\frac{m^{3}}{kg}$)")
ax.set_title("Ejemplo de un gráfico")
ax.legend()
ax.legend(loc='lower right', shadow=True) # locdefine las posiciones

plt.savefig('EjemploFigura.png',dpi=300)
````

### NOTAS

Las ecuaciones que llevan "\" por ejemplo "\frac" necesitan llevar doble diagonal, ejemplo:
```python
ax.set_ylabel("y values ($\frac{m^{3}}{kg}$)")
```

debe de ser:

```python
ax.set_ylabel("y values ($\\frac{m^{3}}{kg}$)")
```
De otra manera mandara error cuando intente compilar la imagen.
